```python
diamonds['volume']=diamonds.apply(lambda row:row['x']*row['y']*row['z'] if row['depth']>60 else 8,axis=1)
diamonds.head(20)
```

```python
diamonds['price'].isnull().sum()
mean_price=diamonds['price'].mean()
diamonds['price']=diamonds['price'].fillna(mean_price)
diamonds.head(20)
```
